###**Nube de Palabras**
Decidimos comenzar el estudio de la información con la que contamos utilizando alguna técnica que, de manera instántanea y general, nos muestre datos que sirvan de base para el planteo de la estrategia de análisis. Para esto nos inclinamos en realizar Nubes de Palabras, la cual es una herramienta muy utilizada ultimamente en la comunicación.

Definimos una función para armar dicha Nube ya que la utilizaremos de manera repetida. La misma utiliza la librería Word Cloud.


```
def NubePalabras(df_filtro,colFiltro):
  
  ListaFiltro=df_filtro[colFiltro].tolist()
  stopwords = []                                            # Variable utilizada para ignorar palabras en el algoritmo.
                                                            # La anulamos para que tome todas las palabras
  texto=''

  for filtro in ListaFiltro:                                # Tomamos solo el nombre de los paises eliminando lo que pudiera contener los ()
    try:                                                    # Adicionalmente, si el pais esta formado con palabras compuestas, 
      filtro=filtro[:filtro.index('(')]                     # juntamos las palabras para que sean tomadas como una sola
      filtro=filtro.replace(" ","")
    except:
      filtro=filtro.replace(" ","")
    texto += filtro + " " 

  wordcloud = WordCloud(width = 600, height = 600,          # Se genera el algoritmo utilizando la libreria Word Cloud
                  background_color ='white',
                  collocations=False,
                  stopwords = stopwords,
                  min_font_size = 4).generate(texto)
  
  return(wordcloud)
```

### **Cantidad de ataques**

Se grafica la cantidad de ataques a lo largo del tiempo. Se dividen los ataques en 4 categorías (índice de gravedad):
*   Sin víctimas fatales
*   Con 1 a 10 muertos
*   Con 11 a 50 muertos
*   Con más de 50 muertos

Partimos de la base que con el paso de los años los países adquieren mayor desarrollo (tecnología, conocimiento, etc.) y esto hace que los ataques disminuyan con el tiempo. Además sospechamos que los ataques "leves" (primeras dos categorías) son mucho más frecuentes que los ataques "graves" (con muchas víctimas).




```
df_victims = df_init_clean.loc[df_init_clean.nkill >= 0]

df_attack0 = df_victims[df_victims['nkill'] == 0]['eventid'].groupby([df_victims['iyear']]).count()
df_attack1 = df_victims[(df_victims['nkill'] >= 1) & (df_victims['nkill'] <= 10)]['eventid'].groupby([df_victims['iyear']]).count()
df_attack2 = df_victims[(df_victims['nkill'] >= 11) & (df_victims['nkill'] <= 50)]['eventid'].groupby([df_victims['iyear']]).count()
df_attack3 = df_victims[df_victims['nkill'] >= 51]['eventid'].groupby([df_victims['iyear']]).count()

fig, ax = plt.subplots(figsize=(12, 8))
ax.plot(df_attack0, color='black', label='Sin víctimas', linewidth=2)
ax.plot(df_attack1, color='blue', label='1 a 10 muertos', linewidth=2)
ax.plot(df_attack2, color='gray', label='11 a 50 muertos', linewidth=2)
ax.plot(df_attack3, color='red', label='Más de 50 muertos', linewidth=2)

ax.tick_params(axis='x', labelsize=12)
ax.tick_params(axis='y', labelsize=12)
ax.grid(visible=True, linestyle='-', linewidth=0.5)
ax.set_xlim([1970, 2017])
ax.set_ylim([0, 8000])
ax.xaxis.set_major_locator(plt.MaxNLocator(10))

ax.set_title('Cantidad de Ataques', loc = "Center", fontdict = {'fontsize':18, 'fontweight':'bold', 'color':'black'})
ax.legend(loc='upper left', frameon=True, fontsize=14)
plt.show
```

**Conclusión**

Contrariamente a lo supuesto la cantidad de ataques ha ido en ascenso con el tiempo y se ha incrementado notablemente en la última década.

Por otro lado sí se verifica nuestra hipótesis inicial que la cantidad de ataques "leves" (sin víctimas fatales o pocas) son muchísimos más frecuentes que los ataques "graves" (con muchas víctimas).

###**Cantidad de ataques Vs % de PBI destinado a sector Militar (2015-2017)**

Partiendo del análisis anterior y limitándonos a los 5 países que presentan más ataques en el intervalo temporal 2015-2017, comparamos ahora la evolución en la cantidad de ataques vs el % del PBI del país destinado al sector militar año por año.

Siendo que el PBI representa el valor de mercado de la producción final de bienes y servicios de un país, el PBI militar implica todos los gastos corrientes y de capital relativos a las fuerzas armadas, incluidas las fuerzas de mantenimiento de la paz, los ministerios de defensa y demás organismos de Gobierno que participan en proyectos de defensa y las actividades en el área militar.

Esperamos encontrar que a mayor % de PBI destinado al sector Militar para un país haya una menor cantidad de ataques en este, debido a que estaría fortaleciendo su defensa y despliegue militar lo cual desincentivaría la ocurrencia de ataques terroristas.

Para realizar esta comparación construimos un DF que contenga los países deseados y agrupamos la cantidad de ataques por cada año.
A su vez, tomamos el DF con datos del PBI Militar anual por país y filtramos la información para obtener el valor en los años a evaluar.
A modo de tener una escala aplicable para amabas variables, procedimos a reescalar el PBI militar x 1000. 

```
# Generación de DataFrame
df_paises_max_ataques = df_ataques_15_17_grouped.head(5)
lista_paises_top_5 = df_paises_max_ataques.index.tolist()
df_top_5_paises = df_ataques_15_17_filtered_2.loc[df_ataques_15_17_filtered_2.Country.isin(lista_paises_top_5)]

# Agrupamiento de datos por país y por año
df_top_5_grouped = df_top_5_paises.groupby(['Country', 'iyear']).count()
df_top_5_grouped.rename(columns = {'eventid': 'count'}, inplace = True)
df_top_5_ataques = pd.DataFrame(df_top_5_grouped.loc[:,'count'])
df_top_5_ataques = df_top_5_ataques.sort_values(by=['Country','iyear'], ascending = True)
df_top_5_ataques = df_top_5_ataques.stack().reset_index()
del df_top_5_ataques['level_2']
df_top_5_ataques.rename(columns = {'iyear': 'year', 0: 'cantidad_ataques'}, inplace = True)
df_top_5_ataques
```

```
# Generación de DataFrame PBI y transformación de la información
# Agrupación de datos por pais y año
df_pbi_grouped = df_filtered_pbi.groupby('Country').max()
df_pbi_grouped_stacked = df_pbi_grouped.stack().reset_index()
df_pbi_grouped_stacked.rename(columns = {'level_1': 'year', 0: 'Military_PBI'}, inplace = True)
df_pbi_grouped_stacked['year'] = pd.to_numeric(df_pbi_grouped_stacked['year'],downcast='integer')
df_pbi_grouped_stacked['Military_PBI'] = round(1000 * (pd.to_numeric(df_pbi_grouped_stacked['Military_PBI'],downcast='integer')), 2)
df_pbi_grouped_stacked.head()
```

Dado que ambos DF (Cantidad de Ataques y PBI Militar) presentan sus datos agrupados por año y por país procedemos a unificarlos para luego comparar la evolución de las variables.

```
# Merge de DataFrames (cantidad de ataques x año + pbi x año)
df_paises_ataques_pbi = pd.merge(df_top_5_ataques, df_pbi_grouped_stacked, how='inner', on = ['Country', 'year'])
df_paises_ataques_pbi = df_paises_ataques_pbi.groupby(['Country', 'year']).max()
df_ataques_pbi = df_paises_ataques_pbi.reset_index()
df_ataques_pbi
```

Realizamos la visualización de las variables utilizando un bucle para obtener un gráfico combinado por cada país, donde las barras muestran la evolución en la cantidad de ataques y las líneas punteadas la evolución del PBI Militar.

```
# Generación de variables útiles para graficar
paises_grafico = df_ataques_pbi['Country'].unique()
años = df_ataques_pbi['year'].unique()
min_año = min(años)
max_año = max(años)
valores_x = np.arange(min_año - 1, max_año + 1 , 1)
max_cant_ataques = max(df_ataques_pbi['Military_PBI'].tolist())

# Generación de gráfico
fig, ax = plt.subplots(1, len(paises_grafico))#, sharey=True)
fig.set_size_inches(20, 5)
plt.subplots_adjust(wspace=0.2, hspace=0.1) 
plt.xticks(rotation = 45)
colores_bar = ['green','blue','orange','red','brown']

# Bucle para graficar país x país
num = 0
for pais in paises_grafico:
  ax[num].set_title(pais, loc = 'center', fontdict = {'fontsize':12, 'fontweight':'bold', 'color':'k'})
  ax[num].tick_params(axis='x', labelrotation = 0)
  ax[num].set_xticks(valores_x, minor=False)
  ax[num].set_ylim([0, max_cant_ataques + 500])
  data1 = df_ataques_pbi[df_ataques_pbi['Country'] == pais]
  data1 = data1.sort_values(by='year')
  ax[num].bar(data1['year'], data1['cantidad_ataques'], color=colores_bar[num], alpha=0.5)
  ax[num].plot(data1['year'], data1['Military_PBI'], color='blue', linestyle='dashed', marker='o', linewidth = 3)
  num = num + 1
```

Para los demás países (Afganistan, India, Pakistan y Filipinas) si bien no comparten el mismo % destinado al PBI Militar, este se mantiene estable o con leves cambios durante el período analizado. A su vez tampoco comparten la misma evolución en cuanto a cantidad de ataques.

En tanto India y Pakistan, así como Afganistan y Filipinas tienen similares valores de porcentajes de PBI M. respectivamente, no sucede lo mismo con la cantidad de ataques, ya que India y Filipinas tienen una cantidad estable y semejante en los años analizados mientras que Afganistan y Pakistan presentan una leve disminución año tras año. 

De esta forma podemos decir que para los países mas atacados entre 2015-2017 la relación que esperabamos tener entre la cantidad de ataques terroristas y el % del PBI Militar no es la que finalmente encontramos en los datos.
